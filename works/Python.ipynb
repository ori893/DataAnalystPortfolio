```python
## import addons 
import seaborn as sb
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
## הגדרת ספריית עבודה
os.chdir('C:\python\Data_sets')
```

```python
## הגדרת דאטאפריים בתור קובץ טקסט
df = pd.read_csv('diamonds.csv')
```

```python
## מציאת מחיר מקסימאלי
df['price'].max()
```

```python
## מחיר ממוצע של יהלום
df['price'].mean()
```

```python
## כמות יהלומים המוגדרים כ"אידיאלים" 
df['cut'].value_counts()['Ideal']
```

```python
## סוגי הצבעים של היהלומים וכמותם
df['color'].value_counts()
```

```python
## החציון קאראט של יהלומים מסוג Premium
df[df['cut'] == 'Premium']['carat'].median()
```

```python
## ממוצע carat לכל סוג cut.
df.groupby('cut')['carat'].mean()
```

```python
##  ממוצע מחיר לכל סוג צבע
df.groupby('color')['price'].mean()
```

```python
##  גרף קרט-מחיר  
plt.plot(df['price'], df['carat'])
plt.xlabel('price')
plt.ylabel('carat')
plt.title('Carat-Price chart')
plt.show()
```

```python
## גרף התפלגות מחירים
df['price'].plot(kind='hist', bins=50, title='Histogram: Price Distribution') 
```

```python
## גרף צבע יהלום אל מול מחיר
sb.kdeplot(data=df, x='price', hue='color')
```
